```
import numpy as np
nsamples=100
f=3
x=np.arange(nsamples)
print(x)
y1 = np.sin(2*np.pi*f*(x/nsamples))
plt.figure()
plt.stem(x,y1,'r',)
plt.plot(x,y1)
plt.xlabel("Time")
plt.ylabel("Aplitude")
plt.show()
```

```
import scipy
def fft_plot(audio, sr):
  n=len(audio)
  F=1/sr
  yf=scipy.fft(audio)
  #for i in range(100):
  #   print(np.abs(yf[i]))
  xf=np.linspace(0.0,1.0/(2.0*F),n//2)
  fig,ax=plt.subplots()
  ax.plot(xf,2.0/n*np.abs(yf[:n//2]))
  plt.grid()
  plt.xlabel("Frequency")
  plt.ylabel("Magnitude")
  return plt.show()
```

```
fft_plot(y1,1)
```

```
f=10
y2 = np.sin(2*np.pi*f*(x/nsamples))
fft_plot(y2,1)
```

```
from librosa import stft

def stft_plot(audio):
  yf=stft(audio)
  #print(yf)
  librosa.display.specshow(librosa.amplitude_to_db(yf,ref=np.max), y_axis='log', x_axis='time')
  plt.title('Power spectrogram')
  plt.colorbar(format='%+2.0f dB')
  plt.tight_layout()
```

```
stft_plot(file1_samples)
```

```
stft_plot(file2_samples)
```
